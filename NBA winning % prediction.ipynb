```python
#first we are creating the variable colname and inserting the object varibale columns in it
# for loop 
colname=[]
for i in df.columns:
    if df[i].dtype=="object":
        colname.append(i)
print(colname)
```

```python
# converting the object(categorical) variable into numerical variable using label encoder
# lablel encoder is used to convert the categorical data into numerical data
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for x in colname:
    df[x]=le.fit_transform(df[x])

# checking the result of labelencoder
df.head()
```

```python
# create X and Y data set
# X = independent variables
# Y = dependent variables

X=df.drop(columns=['win_pct'],axis=1)
Y=df['win_pct']
```

```python
# for proper skewness we have other function 
from scipy.stats import skew


# first we will find skewness in each variable
data_num_skew=X.apply(lambda i:skew(i.dropna()))
print(data_num_skew)

# now subsetting the variables that are highly skewed
# subsetting is required when there are large number of variables present in the data
data_num_skewed=data_num_skew[(data_num_skew>.75) | (data_num_skew<-.75)]
print(data_num_skewed)

# log+1 transformation is necessary if the variables are skewed
# log+1 transformation is to get safe from zeros if present in the data which are skewed over .75

X[data_num_skewed.index]=np.log1p(X[data_num_skewed.index])

# here we are log transforming the skewed variable in the x data set
# .index is used because we want to log transform the skewed variables in a single line of code 

print(data_num_skewed)
print(X)
```

```python
corr_df=X.corr(method="pearson")
print(corr_df)

# after creating and storing the corr values we will use heatmap to visualize the corelation
plt.figure(figsize=(22,22))
sn.heatmap(corr_df,vmax=1.0,vmin=-1.0,annot=True)
plt.show()
```

```python
# 2) Vif ( variance inflation factor) for easy multicolinearity check
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
vif_data=pd.DataFrame()
vif_data["features"]=X.columns
print(vif_data)
vif_data["VIF Factor"]=[vif(X.values,i)for i in range(X.shape[1])]
vif_data.round(2)
```

```python
# Dropping the variables which has high vif values indicating multicollinearity between them
X=X.drop(columns=['TOV','OREB','DREB','REB','attempts_2P','Win','PTS','attempts_3P','FG','FGA'],axis=1)
```

```python
X
```

```python
# splitting the data
from sklearn.model_selection import train_test_split 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
print(X_train)
print(Y_train)
```

```python
# creating the model
from sklearn.linear_model import LinearRegression

# create model object
lm=LinearRegression()

# pass the train data using fit function
lm.fit(X_train,Y_train)

# print intercept and coefficient values
print(lm.intercept_)
print(lm.coef_)

# intercept value is constant beta values
# coeficient values are beta1,beta2,beta3, values
```

```python
# once the model is ready we can pass predict function
Y_pred=lm.predict(X_test)
print(Y_pred)
```

```python
# Evaluation part
# here we will find out r2 , adjustedr2,rmse to check how good the model is

from sklearn.metrics import r2_score,mean_squared_error

r2=r2_score(Y_test,Y_pred)
print(r2)

# r2 score >0.7 then the model is good

adjusted_r_squared=1-(1-r2)*(len(Y)-1)/(len(Y)-X.shape[1]-1)
print(adjusted_r_squared)

rmse=np.sqrt(mean_squared_error(Y_test,Y_pred))
print(rmse)

# rmse value should be close to the min value of the dataset, then the model is good

print(min(Y_test))
```
